```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("summary.csv", encoding='utf-16')

```

```python
print(df.head()) 
```

```python
import matplotlib.pyplot as plt

# Filter out the 'Sequential' row
df_filtered = df[df["Threads"] != "Sequential"]

# Plot speed-up
plt.figure(figsize=(8, 5))
plt.plot(df_filtered["Threads"], df_filtered["Speedup"], marker='o', label='Measured Speed-up')
plt.axhline(1, color='gray', linestyle='--', label='Baseline (1×)')
plt.xlabel("Threads")
plt.ylabel("Speed-up")
plt.title("Speed-up vs Number of Threads")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Example benchmark data (replace with your actual data)
# Number of threads (0 or 1 means sequential)
threads = [1, 2, 4, 8]

# Corresponding execution times in milliseconds (example)
times_ms = [4800, 2500, 1400, 1150]

# Sequential time (assumed first value here)
seq_time = times_ms[0]

# Calculate speedup and efficiency
speedup = [seq_time / t for t in times_ms]
efficiency = [s / n for s, n in zip(speedup, threads)]

# Plot Execution Time
plt.figure(figsize=(10, 6))
plt.plot(threads, times_ms, 'o-', color='blue', label='Execution Time (ms)')
plt.xlabel('Number of Threads')
plt.ylabel('Execution Time (ms)')
plt.title('Execution Time vs Number of Threads')
plt.grid(True)
plt.xticks(threads)
plt.legend()
plt.tight_layout()
plt.savefig('execution_time.png')  # save plot if needed
plt.show()

# Plot Speedup
plt.figure(figsize=(10, 6))
plt.plot(threads, speedup, 'o-', color='green', label='Speedup')
plt.plot(threads, threads, 'k--', label='Ideal Speedup')
plt.xlabel('Number of Threads')
plt.ylabel('Speedup')
plt.title('Speedup vs Number of Threads')
plt.grid(True)
plt.xticks(threads)
plt.legend()
plt.tight_layout()
plt.savefig('speedup.png')
plt.show()

# Plot Efficiency
plt.figure(figsize=(10, 6))
plt.plot(threads, efficiency, 'o-', color='red', label='Efficiency')
plt.xlabel('Number of Threads')
plt.ylabel('Efficiency')
plt.title('Efficiency vs Number of Threads')
plt.grid(True)
plt.xticks(threads)
plt.ylim(0, 1.1)
plt.legend()
plt.tight_layout()
plt.savefig('efficiency.png')
plt.show()
```
