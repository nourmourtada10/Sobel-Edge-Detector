```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("summary.csv", encoding='utf-16')

```

```python
print(df.head()) 
```

```python
import matplotlib.pyplot as plt

# Filter out the 'Sequential' row
df_filtered = df[df["Threads"] != "Sequential"]

# Plot speed-up
plt.figure(figsize=(8, 5))
plt.plot(df_filtered["Threads"], df_filtered["Speedup"], marker='o', label='Measured Speed-up')
plt.axhline(1, color='gray', linestyle='--', label='Baseline (1×)')
plt.xlabel("Threads")
plt.ylabel("Speed-up")
plt.title("Speed-up vs Number of Threads")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample data from your thread sweep results
data = {
    "Threads": ["Sequential", 1, 2, 4, 8],
    "Avg_Time_ms": [4754.99, 4862.17, 2522.87, 1450.41, 1161.20],
    "Speedup": [1.00, 0.98, 1.88, 3.28, 4.09]
}

df = pd.DataFrame(data)
# Convert 'Sequential' to 0 threads for plotting purposes
df["Threads_Num"] = df["Threads"].replace({"Sequential": 0}).astype(int)

# Plot 1: Speed-up vs Number of Threads
plt.figure(figsize=(8,5))
plt.plot(df["Threads_Num"][1:], df["Speedup"][1:], marker='o', linestyle='-', color='b')
plt.title("Speed-up vs Number of Threads")
plt.xlabel("Number of Threads")
plt.ylabel("Speed-up")
plt.xticks(df["Threads_Num"][1:])
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot 2: Average Runtime vs Number of Threads (log scale for better visualization)
plt.figure(figsize=(8,5))
plt.plot(df["Threads_Num"], df["Avg_Time_ms"], marker='o', linestyle='-', color='r')
plt.title("Average Runtime vs Number of Threads")
plt.xlabel("Number of Threads")
plt.ylabel("Average Runtime (ms)")
plt.xticks(df["Threads_Num"], labels=df["Threads"])
plt.yscale('log')
plt.grid(True, which="both", ls="--")
plt.tight_layout()
plt.show()
```
