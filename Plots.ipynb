```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("summary.csv", encoding='utf-16')

```

```python
print(df.head()) 
```

```python
import matplotlib.pyplot as plt

# Filter out the 'Sequential' row
df_filtered = df[df["Threads"] != "Sequential"]

# Plot speed-up
plt.figure(figsize=(8, 5))
plt.plot(df_filtered["Threads"], df_filtered["Speedup"], marker='o', label='Measured Speed-up')
plt.axhline(1, color='gray', linestyle='--', label='Baseline (1Ã—)')
plt.xlabel("Threads")
plt.ylabel("Speed-up")
plt.title("Speed-up vs Number of Threads")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
